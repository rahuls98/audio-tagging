```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
TRAIN = pd.read_csv('fsd_train_targets.csv')
TEST = pd.read_csv('fsd_test_targets.csv')
```

```python
train_set_counts = {}

for i in range(1,81):
    label = TRAIN.columns[i]
    count = TRAIN[label].value_counts()
    train_set_counts[label] = count[1]
    
train_set_counts
```

```python
test_set_counts = {}

for i in range(1,81):
    label = TEST.columns[i]
    count = TEST[label].value_counts()
    test_set_counts[label] = count[1]
    
test_set_counts
```

```python
targets = TRAIN.columns[1:]
targets = list(targets)
```

```python
#Folder creating script

import os

for i in targets:
    path = "./" + i
    try:
        os.mkdir(path)
    except OSError:
        print ("Creation of the directory %s failed" % path)
    else:
        print ("Successfully created the directory %s " % path)
```

```python
TRAIN.loc[TRAIN['Accelerating_and_revving_and_vroom']==1]["fname"].to_list()
```

```python
train_to_move = {}

for i in range(1,81):
    label = TRAIN.columns[i]
    files = TRAIN.loc[TRAIN[label]==1]["fname"].to_list()
    train_to_move[label] = {}
    train_to_move[label]["files"] = files
    train_to_move[label]["count"] = len(files)
    
train_to_move
```

```python
test_to_move = {}

for i in range(1,81):
    label = TEST.columns[i]
    files = TEST.loc[TEST[label]==1]["fname"].to_list()
    test_to_move[label] = {}
    test_to_move[label]["files"] = files
    test_to_move[label]["count"] = len(files)
    
test_to_move
```
