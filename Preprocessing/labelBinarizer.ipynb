```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer 
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/rahuls98/audio-tagging/master/test.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df = df[['fname','labels']]
```

```python
df.shape
```

```python
labels = df.labels
labels
```

```python
labels = labels.to_list()
```

```python
labels = [tuple(x.split(',')) for x in labels]
labels
```

```python
one_hot = MultiLabelBinarizer()
encoding = one_hot.fit_transform(labels)
label_names = one_hot.classes_
```

```python
df2 = pd.DataFrame(encoding)
df2.columns = label_names
```

```python
df2.head()
```

```python
df = df.drop(columns='labels')
df.shape
```

```python
final = pd.concat([df, df2], axis=1)
```

```python
final.shape
```
